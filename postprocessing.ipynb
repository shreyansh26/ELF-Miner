```python
df['ELFVersion'] = df['ELFVersion'].apply(lambda x: x[2:])
df['Flags'] = df['Flags'].apply(lambda x: x[2:])
```

```python
df['EntryPointAddress'] = df['EntryPointAddress'].apply(lambda x: x[2:])
```

```python
df.head()
```

```python
req_col = [a for a in df.columns if '_size' in a]
req_col
```

```python
for i in req_col:
    df[i] = df[i].apply(lambda x: str(x))
    df[i] = df[i].apply(lambda x: x.replace(".0", ''))
for i in req_col:
    df[i] = df[i].apply(lambda x: int(x, 16) if x != "nan" else x)
```

```python
df.head()
```

```python
df.to_csv("results_final.csv", index=False)
```
